```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/intel/Desktop/Middle-east-project/data/Middle_East_Economic_Data_1990_2024_with_Oil.csv")

df.head()
```

```python
df=df.fillna(df.median(numeric_only=True))
```

```python
df
```

```python
df=df.dropna(subset=["Country"])
```

```python
df.duplicated().sum()
```

```python
gdp_country=df.groupby("Country")["GDP_growth_annual_pct"].mean()

gdp_country.sort_values(ascending=False)
```

```python
gdp_country.sort_values().plot(kind="barh", figsize=(8,5))

plt.title("Average GDP Growth by Country")
plt.xlabel("GDP Growth (%)")
plt.ylabel("Country")


plt.savefig("C:/Users/intel/Desktop/Middle-east-project/visuals/gdp_growth_country.png")

plt.show()
```

## insights
- Qatar and UAE show higher average GDP growth compared to other Middle Eastern countries.

- Countries with strong oil exports tend to experience higher economic growth.

- Some countries show lower growth due to economic diversification challenges.

```python
oil_trend = df.groupby("Year")["Brent_Oil_Price_USD_per_barrel"].first()

oil_trend.head()
```

```python
oil_trend.plot(figsize=(10,5))

plt.title("Brent Oil Price Trend (1990–2024)")
plt.xlabel("Year")
plt.ylabel("Oil Price")

plt.savefig("C:/Users/intel/Desktop/Middle-east-project/visuals/oil_price_trend.png")
plt.show()
```

### Insight

-Oil prices experienced major spikes around 2008 and 2022.

-These spikes coincide with global economic events such as the financial crisis and geopolitical tensions.

-Oil prices are highly volatile, which can impact oil-dependent economies.

```python
plt.scatter(df["Brent_Oil_Price_USD_per_barrel"],
            df["GDP_growth_annual_pct"])

plt.xlabel("Oil Price")
plt.ylabel("GDP Growth (%)")
plt.title("Oil Price vs GDP Growth")


plt.savefig("C:/Users/intel/Desktop/Middle-east-project/visuals/oil_vs_gdp.png")
plt.show()
```

### Insight

-There is a moderate positive relationship between oil prices and GDP growth.

-Higher oil prices generally benefit oil-exporting countries.

-However, the relationship is not perfectly linear because other economic factors also influence growth.

```python
unemployment_country = df.groupby("Country")["Unemployment_total_pct"].mean()

unemployment_country.sort_values(ascending=False)
```

```python
unemployment_country.sort_values().plot(kind="barh", figsize=(8,5))

plt.title("Average Unemployment Rate by Country")
plt.xlabel("Unemployment (%)")
plt.ylabel("Country")



plt.savefig("C:/Users/intel/Desktop/Middle-east-project/visuals/unemployment_country.png")
plt.show()
```

### Insight

-Some countries maintain higher unemployment despite economic growth.

-This may indicate structural issues in the labor market.

-Diversification away from oil sectors could impact employment patterns.
